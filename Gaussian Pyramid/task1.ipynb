```python
import matplotlib.pyplot as plt
from skimage import data, img_as_float
from skimage.transform import pyramid_gaussian

# Load an example image
image = img_as_float(data.astronaut())  # RGB image

# Build the Gaussian pyramid
pyramid = tuple(
    pyramid_gaussian(image, max_layer=4, downscale=2, channel_axis=-1)
)

# Print info about each level
print("Gaussian Pyramid Levels:")
for i, layer in enumerate(pyramid):
    print(f"Level {i}: shape = {layer.shape}")

# Plot the pyramid
fig, axes = plt.subplots(1, len(pyramid), figsize=(15, 5))
for i, (ax, layer) in enumerate(zip(axes, pyramid)):
    ax.imshow(layer)
    ax.set_title(f"Level {i}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```

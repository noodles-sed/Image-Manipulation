```python
# --- Pencil Sketch Effect using OpenCV ---
import cv2
import numpy as np
from skimage import data, img_as_ubyte
import matplotlib.pyplot as plt

# Load a sample image (or use a local one)
image = img_as_ubyte(data.astronaut())
original_image = image.copy()

# Convert to grayscale
gray = cv2.cvtColor(original_image, cv2.COLOR_RGB2GRAY)

# Invert and blur the grayscale image
inverted = 255 - gray
blurred = cv2.GaussianBlur(inverted, (21, 21), 0)

# Create pencil sketch
sketch_gray = cv2.divide(gray, 255 - blurred, scale=256)

# Optional: create a color pencil sketch using OpenCV's built-in function
_, sketch_color = cv2.pencilSketch(original_image, sigma_s=60, sigma_r=0.07, shade_factor=0.05)

# Display results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(original_image)
axes[0].set_title("Original Image")
axes[0].axis("off")

axes[1].imshow(sketch_gray, cmap='gray')
axes[1].set_title("Grayscale Pencil Sketch")
axes[1].axis("off")

axes[2].imshow(sketch_color)
axes[2].set_title("Color Pencil Sketch")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
